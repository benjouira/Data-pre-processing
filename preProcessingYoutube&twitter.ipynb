```python
import pandas as pd
import nltk

df = pd.read_csv("https://query.data.world/s/lwoyl5xyoq6sugmiz7wrqf3eb5vckg")
```

```python
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
# first loop the customer reviews field for each product in all the dataset
for i in range(0,df['customer_reviews'].shape[0]):
  # for each product all the review are in the same text separated with "|"
  # so we need to spilt it to get every review a part
  reviews = df['customer_reviews'][i].split('|')
  # after spliting we notice that every single review contains different information separated with "//"
  rvnb = 0
  for review in reviews:
    rvnb = rvnb+1
    rv = review.split('//')
    review_tilte=rv[0]
    rate=rv[1]
    review_date=rv[2]
    customer_info=rv[3]
    comment=rv[4]
    # after spliting we get the review_tilte , review_rate, review_date, customer_info  , comment
    # we notice that customer_info field contains different information separated by "\n"
    cst_info = customer_info.split('\n')
    # the name is in the third position
    customer_name = cst_info[2]
    # .strip() is like .trim() i use it to delete spaces befor and after text
    print("product_name :",df['product_name'][i])
    print("Review number:",rvnb)
    print("Title :",review_tilte.strip())
    print("Rate :",rate.strip())
    print("review date :",review_date.strip())
    print("customer name :",customer_name.strip())
    print("comment :",comment.strip())
    print("******************************************************")
    
```
